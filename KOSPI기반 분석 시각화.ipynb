### 데이터 수가 너무 많이 연도별로 그룹화하여 평균낸 데이터를 추출하고자 함

```python
idx = list(kospi['Date'])
idx
```

```python
for i in range(len(list(kospi['Date']))):
    idx[i] = list(kospi['Date'])[i][:4]
```

```python
idx
```

```python
kospi['Date'] = idx
kospi
```

```python
# 컬럼명을 모두 소문자로 변경
kospi.columns = kospi.columns.str.lower()   
kospi
```

```python
kospi['date']
idxlst = list(kospi['date'])
idxlst
```

```python
idxlst = set(idxlst)   # 중복 제거를 위해 set으로 타입변경 후 
idxlst = list(idxlst)   # 다시 리스트 형태로 변경하여 
idxlst = sorted(idxlst)  # 리스트 정렬

idxlst, len(idxlst)
```

```python
kospi_year = kospi.groupby('date').mean()
kospi_year
```

```python
# 1981년과 2022년 데이터 삭제 (1년치 정보가 충분히 들어있지 않음)
kospi_year = kospi_year[1:-1]
kospi_year
```

### 시가-종가 그래프

```python
import matplotlib.pyplot as plt

# 그림 사이즈 지정 (가로 45인치 * 세로 15인치)
fig = plt.figure(figsize = (80, 30))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

# 종가 그래프
ax1.plot(kospi_year['close'], marker = 'o', markerfacecolor = 'green',
       markersize = 15, color = 'olive', linewidth = 10, label = '종가')
# 시가 그래프
ax2.plot(kospi_year['open'], marker = 'o', markerfacecolor = 'orange',
       markersize = 15, color = 'orange', linewidth = 10, label = '시가')

# 범례 표시
ax1.legend(loc = 'best', fontsize = 50)
ax2.legend(loc = 'best', fontsize = 50)

# 차트 제목
ax1.set_title('1981~2022 종가 그래프[연도별]', size = 70)
ax2.set_title('1981~2022 시가 그래프[연도별]', size = 70)

# 축 이름 
ax1.set_xlabel('date', size= 50)
ax1.set_ylabel('price', size = 50)
ax2.set_xlabel('date', size= 50)
ax2.set_ylabel('price', size = 50)

# 축 눈금 라벨 지정 및 90도 회전
ax1.set_xticklabels(kospi_year.index, rotation=90)
ax2.set_xticklabels(kospi_year.index, rotation=90)

# 축 눈금 라벨 크기
ax1.tick_params(axis="x", labelsize=40)
ax2.tick_params(axis="x", labelsize=40)

plt.show()
```

### 고가-저가 그래프

```python
# 그림 사이즈 지정 (가로 45인치 * 세로 15인치)
fig = plt.figure(figsize = (80, 30))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

# 종가 그래프
ax1.plot(kospi_year['high'], marker = 'o', markerfacecolor = 'red',
       markersize = 15, color = 'red', linewidth = 10, label = '고가')
# 시가 그래프
ax2.plot(kospi_year['low'], marker = 'o', markerfacecolor = 'blue',
       markersize = 15, color = 'blue', linewidth = 10, label = '저가')

# 범례 표시
ax1.legend(loc = 'best', fontsize = 50)
ax2.legend(loc = 'best', fontsize = 50)

# 차트 제목
ax1.set_title('1981~2022 고가 그래프[연도별]', size = 70)
ax2.set_title('1981~2022 저가 그래프[연도별]', size = 70)

# 축 이름 
ax1.set_xlabel('date', size= 50)
ax1.set_ylabel('price', size = 50)
ax2.set_xlabel('date', size= 50)
ax2.set_ylabel('price', size = 50)

# 축 눈금 라벨 지정 및 90도 회전
ax1.set_xticklabels(kospi_year.index, rotation=90)
ax2.set_xticklabels(kospi_year.index, rotation=90)

# 축 눈금 라벨 크기
ax1.tick_params(axis="x", labelsize=40)
ax2.tick_params(axis="x", labelsize=40)

plt.show()
```

### 거래량&변동률 그래프

```python
# 거래량 - 막대그래프
# 전년대비 변동률 - 선그래프
```

```python
ax1 = kospi_year['volume'].plot(kind = 'bar', figsize = (20, 10), width = 0.7)
ax2 = ax1.twinx()  # x축 공유, y축 따로
ax2.plot(kospi_year.index, kospi_year['change'], ls = '--', marker = 'o', markersize = 10,
        color = 'gray', label = '전년대비 변동률(%)')

ax1.set_xlabel('year', size = 20)
ax1.set_ylabel('거래량')
ax2.set_ylabel('전년대비 변동율(%)')

plt.title('1982 - 2020 KOSPI 거래량', size = 30)
ax1.legend(loc = 'best')

plt.show()
```

### 다중선형회귀-Mean 값 예측

#### 1. 데이터세트 분리

```python
from sklearn.model_selection import train_test_split
x = df[['Close', 'Open', 'Volume']]
y = df[['Mean']]

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.8)
```

#### 2. 모델 생성

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```
